```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
region_number = ['ONE', 'TWO', 'THREE', 'FOUR', 'FIVE']
region_sales = [47673, 56560, 56503, 51680, 52543]
```

```python
plt.title('ICE CREAM TRENDS IN FIVE REGIONS IN YEAR 2010')
plt.xlabel('REGION NUMBER')
plt.ylabel('NUMBER OF ICE CREAMS SOLD')
```

```python
x_pos = np.arange(len(region_number))
plt.xticks(x_pos, region_number)
```

```python
# function to put labels on each of the graphs
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2., 1.0 * height,
                '%d' % int(height), ha='center', va='bottom')
```

```python
autolabel(plt.bar(x_pos, region_sales, label='ICE CREAM SALES FOR FIVE REGIONS'))
plt.legend()
plt.show()
```
