```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = ['2010', '2011', '2012']
year_sales = [10981, 10533, 12031]
```

```python
plt.title('ICE CREAM SALES FOR YEAR 2010 - 2012')
plt.xlabel('YEAR')
plt.ylabel('NUMBER OF ICE CREAMS SOLD')
```

```python
x_pos = np.arange(len(years))
plt.xticks(x_pos, years)
```

```python
# function to put labels on each of the graphs
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2., 1.0 * height,
                '%d' % int(height), ha='center', va='bottom')
```

```python
autolabel(plt.bar(x_pos, year_sales, label='ICE CREAM SALES FOR ONE STORE'))
plt.legend()
plt.show()
```

```python
autolabel(plt.bar(x_pos, year_sales, label='ICE CREAM SALES FOR ONE STORE'))
plt.legend(fontsize='large', shadow=True)
plt.show()
```
