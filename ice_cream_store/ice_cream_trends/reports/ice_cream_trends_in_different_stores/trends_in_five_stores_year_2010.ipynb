```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
store_number = ['ONE', 'TWO', 'THREE', 'FOUR', 'FIVE']
store_sales = [11523, 10729, 10791, 10248, 11453]
```

```python
plt.title('ICE CREAM TRENDS IN FIVE STORES IN YEAR 2010')
plt.xlabel('STORE NUMBER')
plt.ylabel('NUMBER OF ICE CREAMS SOLD')
```

```python
x_pos = np.arange(len(store_number))
plt.xticks(x_pos, store_number)
```

```python
# function to put labels on each of the graphs
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2., 1.0 * height,
                '%d' % int(height), ha='center', va='bottom')
```

```python
autolabel(plt.bar(x_pos, store_sales, label='ICE CREAM SALES FOR FIVE STORES'))
plt.legend()
plt.show()
```
