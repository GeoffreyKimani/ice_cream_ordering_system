```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp_range = ['0-7', '8-15', '16-19', '20-23', '24-27', '28-30', '31-33', '34-35']
temp_sales = [2, 10, 17, 20, 34, 40, 63, 68]
```

```python
plt.title('ICE CREAM SALES AGAINST TEMPERATURE IN A STORE')
plt.xlabel('TEMP RANGE')
plt.ylabel('NUMBER OF ICE CREAMS SOLD')
```

```python
x_pos = np.arange(len(temp_range))
plt.xticks(x_pos, temp_range)
```

```python
# function to put labels on each of the graphs
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2., 1.0 * height,
                '%d' % int(height), ha='center', va='bottom')
```

```python
autolabel(plt.bar(x_pos, temp_sales, label='ICE CREAM SALES AGAINST TEMP'))
plt.legend()
plt.show()
```
